```python
from openai import OpenAI
import os

# 从环境变量获取API密钥，如果没有设置则使用默认值
api_key = os.getenv("ALIBABA_API_KEY", "sk-XXXXXXX(自己申请的API_KEY)")

client = OpenAI(
    api_key=api_key,
    base_url="https://dashscope.aliyuncs.com/compatible-mode/v1",
)

unfinished_code = """
def quick_sort(arr):
    ...

arr = [3, 6, 8, 10, 1, 2, 1]
quick_sort(arr)
print(arr)
"""

prompt, suffix = unfinished_code.split("...", 1)

# 构建完整的提示
full_prompt = f"{prompt}请补全这个快速排序函数的实现{suffix}"

try:
    response = client.chat.completions.create(
        model="qwen-plus",
        messages=[
            {"role": "system", "content": "You are a helpful coding assistant. Please complete the code implementation."},
            {"role": "user", "content": full_prompt},
        ],
        max_tokens=500,
    )

    print("\033[1;31m------------------------- Code Completion -------------------------\033[0m")
    print(response.choices[0].message.content)  # 补全的内容

    print("\033[1;31m------------------------- Completed Code -------------------------\033[0m")
    # 将补全的内容插入到原代码中
    completed_code = unfinished_code.replace("...", response.choices[0].message.content)
    print(completed_code)

except Exception as e:
    print(f"\033[1;31m错误：{e}\033[0m")
    print("\n可能的解决方案：")
    print("1. 检查API密钥是否正确")
    print("2. 检查网络连接")
    print("3. 检查API端点是否可访问")
    print("4. 检查API配额是否充足")
```

```python
#多轮对话模式
from openai import OpenAI
import os

api_key = os.getenv("ALIBABA_API_KEY", "sk-XXXXXXX(自己申请的API_KEY")

client = OpenAI(
    api_key=api_key,
    base_url="https://dashscope.aliyuncs.com/compatible-mode/v1",
)

messages = [
    {"role": "system", "content": "You are a helpful assistant."}
]

print("\033[1;32m=== 欢迎使用AI聊天助手 ===\033[0m")
print("\033[1;33m输入 'exit' 或 'quit' 结束对话\033[0m")
print("\033[1;33m输入 'clear' 清空对话历史\033[0m")
print("\033[1;33m输入 'help' 查看帮助信息\033[0m")
print("\033[1;32m" + "="*30 + "\033[0m\n")

while True:
    user_input = input("\033[1;34mYou: \033[0m").strip()

    # 检查退出命令
    if user_input.lower() in ['exit', 'quit', '退出', '结束']:
        print("\033[1;32mAssistant: \033[0m再见！感谢使用AI聊天助手。")
        print("\033[1;32m" + "="*30 + "\033[0m")
        break

    # 检查清空历史命令
    if user_input.lower() in ['clear', '清空', '清除']:
        messages = [{"role": "system", "content": "You are a helpful assistant."}]
        print("\033[1;33m对话历史已清空\033[0m\n")
        continue

    # 检查帮助命令
    if user_input.lower() in ['help', '帮助', '?']:
        print("\033[1;33m帮助信息：\033[0m")
        print("- 输入 'exit' 或 'quit' 结束对话")
        print("- 输入 'clear' 清空对话历史")
        print("- 输入 'help' 查看此帮助信息")
        print("- 其他任何输入将作为对话内容发送给AI\n")
        continue

    # 跳过空输入
    if not user_input:
        continue

    # 添加用户消息到对话历史
    messages.append({"role": "user", "content": user_input})

    try:
        # 调用API获取响应
        response = client.chat.completions.create(
            model="qwen-plus",
            messages=messages,
            stream=True,
            max_tokens=1000
        )

        # 收集AI的完整响应
        ai_response = ""
        print("\033[1;32mAssistant: \033[0m", end="")

        for chunk in response:
            if chunk.choices[0].delta.content:
                content = chunk.choices[0].delta.content
                ai_response += content
                print(content, end="", flush=True)

        print()  # 换行

        # 将AI响应添加到对话历史
        messages.append({"role": "assistant", "content": ai_response})

        # 显示对话结束标记
        print("\033[1;30m" + "-"*50 + "\033[0m")

    except Exception as e:
        print(f"\033[1;31m错误：{e}\033[0m")
        print("请检查API密钥和网络连接")
        # 移除失败的用户消息
        messages.pop()
        continue

    print()  # 对话间隔
```

2. 提示工程

 2.1 思维链（chain of thought）
 CoT是一种改进的提示工程技术，目的在于提升大模型在复杂推理任务上的表现，对于复杂问题尤其是复杂的数学题大模型很难直接给出正确答案。如算术推理、常识推理、符号推理等。CoT通过要求模型在输出最终答案之前，显示输出中间逐步的推理步骤这一方法来增强大模型的算数、常识和推理能力。

 2.2 Few-Shot-CoT提示词
 Few-Shot-CoT是一种在提示大型语言模型时，通过提供少量示例，展示逐步推理过程，以提升模型在复杂任务上的表现的方法。Few-Shot CoT通过在提示中加入示例，指导模型生成类似的推理步骤，从而提高其在复杂任务上的准确性。

 系统提示词：You are a helpful assistant.
 用户提示词：
 问题：单词 Banana 中一共有多少个 n？
 答案：单词 "Banana" 中一共有多少个 "n"，我们可以逐步分析：
 1. 首先，写出单词 "Banana"：
 -	B
 -	a
 -	n
 -	a
 -	n
 -	a
 2. 然后，数一数有多少个 "n"：
 -	第三个字母是 "n"
 -	第五个字母是 "n"
 3. 总结一下，单词 "Banana" 中一共有 2 个 "n"。

```python
#放入函数调用的形式实现
from openai import OpenAI
import os

api_key = os.getenv("ALIBABA_API_KEY", "sk-XXXXXXX(自己申请的API_KEY")

client = OpenAI(
    api_key=api_key,
    base_url="https://dashscope.aliyuncs.com/compatible-mode/v1",
)
def request_with_standard(question):
    print("\033[1;31m------------------------- Standard -------------------------\033[0m")
    print(f"\033[1;33mQuestion: {question}\033[0m")
    response = client.chat.completions.create(messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": question},
    ], model="qwen-plus")
    response_text = response.choices[0].message.content
    print(f"\033[1;32mResponse: {response_text}\033[0m")
    return response_text

if __name__ == "__main__":
    question1 = "单词 Strawberry 中一共有多少个 r？"
    request_with_standard(question1)

    question2 = "小明早上 8:00 离开家，以每小时 40 公里的速度行驶。早上 8:30，他停下来 10 分钟加油。然后他以每小时 60 公里的速度行驶。如果目的地距离 120 公里，他什么时候到达目的地？"
    request_with_standard(question2)
```

3. 工具/函数调用API

工具调用(Function Calling)功能允许LLM在生成响应时调用外部函数或API，以获取实时数据或执行特定操作。通过定义函数的名称、参数和描述，模型可以根据用户输入智能地决定是否调用相应地函数，并生成符合函数签名的JSON对象。
