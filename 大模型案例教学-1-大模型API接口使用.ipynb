```python
from openai import OpenAI
import os

# ä»ç¯å¢ƒå˜é‡è·å–APIå¯†é’¥ï¼Œå¦‚æœæ²¡æœ‰è®¾ç½®åˆ™ä½¿ç”¨é»˜è®¤å€¼
api_key = os.getenv("ALIBABA_API_KEY", "sk-XXXXXXX(è‡ªå·±ç”³è¯·çš„API_KEY)")

client = OpenAI(
    api_key=api_key,
    base_url="https://dashscope.aliyuncs.com/compatible-mode/v1",
)

unfinished_code = """
def quick_sort(arr):
    ...

arr = [3, 6, 8, 10, 1, 2, 1]
quick_sort(arr)
print(arr)
"""

prompt, suffix = unfinished_code.split("...", 1)

# æ„å»ºå®Œæ•´çš„æç¤º
full_prompt = f"{prompt}è¯·è¡¥å…¨è¿™ä¸ªå¿«é€Ÿæ’åºå‡½æ•°çš„å®ç°{suffix}"

try:
    response = client.chat.completions.create(
        model="qwen-plus",
        messages=[
            {"role": "system", "content": "You are a helpful coding assistant. Please complete the code implementation."},
            {"role": "user", "content": full_prompt},
        ],
        max_tokens=500,
    )

    print("\033[1;31m------------------------- Code Completion -------------------------\033[0m")
    print(response.choices[0].message.content)  # è¡¥å…¨çš„å†…å®¹

    print("\033[1;31m------------------------- Completed Code -------------------------\033[0m")
    # å°†è¡¥å…¨çš„å†…å®¹æ’å…¥åˆ°åŸä»£ç ä¸­
    completed_code = unfinished_code.replace("...", response.choices[0].message.content)
    print(completed_code)

except Exception as e:
    print(f"\033[1;31mé”™è¯¯ï¼š{e}\033[0m")
    print("\nå¯èƒ½çš„è§£å†³æ–¹æ¡ˆï¼š")
    print("1. æ£€æŸ¥APIå¯†é’¥æ˜¯å¦æ­£ç¡®")
    print("2. æ£€æŸ¥ç½‘ç»œè¿æ¥")
    print("3. æ£€æŸ¥APIç«¯ç‚¹æ˜¯å¦å¯è®¿é—®")
    print("4. æ£€æŸ¥APIé…é¢æ˜¯å¦å……è¶³")
```

```python
#å¤šè½®å¯¹è¯æ¨¡å¼
from openai import OpenAI
import os

api_key = os.getenv("ALIBABA_API_KEY", "sk-XXXXXXX(è‡ªå·±ç”³è¯·çš„API_KEY")

client = OpenAI(
    api_key=api_key,
    base_url="https://dashscope.aliyuncs.com/compatible-mode/v1",
)

messages = [
    {"role": "system", "content": "You are a helpful assistant."}
]

print("\033[1;32m=== æ¬¢è¿ä½¿ç”¨AIèŠå¤©åŠ©æ‰‹ ===\033[0m")
print("\033[1;33mè¾“å…¥ 'exit' æˆ– 'quit' ç»“æŸå¯¹è¯\033[0m")
print("\033[1;33mè¾“å…¥ 'clear' æ¸…ç©ºå¯¹è¯å†å²\033[0m")
print("\033[1;33mè¾“å…¥ 'help' æŸ¥çœ‹å¸®åŠ©ä¿¡æ¯\033[0m")
print("\033[1;32m" + "="*30 + "\033[0m\n")

while True:
    user_input = input("\033[1;34mYou: \033[0m").strip()

    # æ£€æŸ¥é€€å‡ºå‘½ä»¤
    if user_input.lower() in ['exit', 'quit', 'é€€å‡º', 'ç»“æŸ']:
        print("\033[1;32mAssistant: \033[0må†è§ï¼æ„Ÿè°¢ä½¿ç”¨AIèŠå¤©åŠ©æ‰‹ã€‚")
        print("\033[1;32m" + "="*30 + "\033[0m")
        break

    # æ£€æŸ¥æ¸…ç©ºå†å²å‘½ä»¤
    if user_input.lower() in ['clear', 'æ¸…ç©º', 'æ¸…é™¤']:
        messages = [{"role": "system", "content": "You are a helpful assistant."}]
        print("\033[1;33må¯¹è¯å†å²å·²æ¸…ç©º\033[0m\n")
        continue

    # æ£€æŸ¥å¸®åŠ©å‘½ä»¤
    if user_input.lower() in ['help', 'å¸®åŠ©', '?']:
        print("\033[1;33må¸®åŠ©ä¿¡æ¯ï¼š\033[0m")
        print("- è¾“å…¥ 'exit' æˆ– 'quit' ç»“æŸå¯¹è¯")
        print("- è¾“å…¥ 'clear' æ¸…ç©ºå¯¹è¯å†å²")
        print("- è¾“å…¥ 'help' æŸ¥çœ‹æ­¤å¸®åŠ©ä¿¡æ¯")
        print("- å…¶ä»–ä»»ä½•è¾“å…¥å°†ä½œä¸ºå¯¹è¯å†…å®¹å‘é€ç»™AI\n")
        continue

    # è·³è¿‡ç©ºè¾“å…¥
    if not user_input:
        continue

    # æ·»åŠ ç”¨æˆ·æ¶ˆæ¯åˆ°å¯¹è¯å†å²
    messages.append({"role": "user", "content": user_input})

    try:
        # è°ƒç”¨APIè·å–å“åº”
        response = client.chat.completions.create(
            model="qwen-plus",
            messages=messages,
            stream=True,
            max_tokens=1000
        )

        # æ”¶é›†AIçš„å®Œæ•´å“åº”
        ai_response = ""
        print("\033[1;32mAssistant: \033[0m", end="")

        for chunk in response:
            if chunk.choices[0].delta.content:
                content = chunk.choices[0].delta.content
                ai_response += content
                print(content, end="", flush=True)

        print()  # æ¢è¡Œ

        # å°†AIå“åº”æ·»åŠ åˆ°å¯¹è¯å†å²
        messages.append({"role": "assistant", "content": ai_response})

        # æ˜¾ç¤ºå¯¹è¯ç»“æŸæ ‡è®°
        print("\033[1;30m" + "-"*50 + "\033[0m")

    except Exception as e:
        print(f"\033[1;31mé”™è¯¯ï¼š{e}\033[0m")
        print("è¯·æ£€æŸ¥APIå¯†é’¥å’Œç½‘ç»œè¿æ¥")
        # ç§»é™¤å¤±è´¥çš„ç”¨æˆ·æ¶ˆæ¯
        messages.pop()
        continue

    print()  # å¯¹è¯é—´éš”
```
